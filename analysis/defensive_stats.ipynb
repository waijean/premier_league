```python
import pandas as pd
```

```python
df = pd.read_csv("../data/defense_stats.csv", header=[0,1])
```

```python
df.head()
```

```python
col_list = [b if a.startswith("Unnamed") else a + " " + b for a,b in df.columns.to_list()]
df.columns = col_list
```

```python
def pre_process(df):
    df = df.copy()
    df = df.drop(columns=["Rk", "Matches"])
    df = df[df.Player != "Player"]
    dtype_dict = {col:float for col in df.columns[6:]}
    df = df.astype(dtype_dict, copy=False)
    return df

processed_df = pre_process(df)
```

```python
processed_df.info() 
```

```python
defenders_df = processed_df.loc[df["Pos"] == "DF"]
def_stats_df = defenders_df[["Player", "Squad", "Age", "90s", "Tackles Tkl", "Blocks Blocks", "Int", "Tkl+Int", "Clr", "Err"]]
```

```python
def_stats_df.loc[def_stats_df["90s"] >= 5].sort_values(by="Tackles Tkl", ascending=False).head(10)
```

```python
def_stats_df.loc[def_stats_df["90s"] >= 5].sort_values(by="Clr", ascending=False).head(10)
```

```python
def_stats_df.loc[def_stats_df["90s"] >= 5].sort_values(by="Blocks Blocks", ascending=False).head(10)
```

```python
def_stats_df.loc[def_stats_df["90s"] >= 5].sort_values(by="Int", ascending=False).head(10)
```

```python
def_stats_df["Int+Clr"] = def_stats_df["Int"] + def_stats_df["Clr"]
def_stats_df.loc[def_stats_df["90s"] >= 5].sort_values(by="Int+Clr", ascending=False).head(10)
```
